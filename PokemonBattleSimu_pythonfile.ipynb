```python
def first_turn():
  if player1[5] > player2[5]: # checks the speed stat of two pokemons and returns the faster one
    print(f'{player1[0]} is faster than {player2[0]}, so player_1 moves first.')
    return player1
  else:
    print(f'{player2[0]} is faster than {player1[0]}, so player_2 moves first.')
    return player2  # Return the player who goes first

first_player = first_turn() # Store the player who goes first
```

```python
def Type_multiplier(attacker_type, defender_type):

    # Super effective cases
    if (attacker_type == "Fighting" and defender_type == "Normal") or \
       (attacker_type == "Ground" and defender_type in ["Electric", "Fire"]):
        return 2
    # Not very effective cases
    elif (attacker_type == "Fire" and defender_type == "Fire") or \
         (attacker_type == "Electric" and defender_type == "Electric"):
        return 0.5
    #no effect case
    elif attacker_type == "Electric" and defender_type == "Ground":
        return 0
    # All other matchups
    else:
        return 1
```

```python
def attack(player):
  global p1_HP, p2_HP

  #print(multiplier)
  if player == player1:
    multiplier = Type_multiplier(player1[1],player2[1])

    print('multi : ',multiplier)

    p2_HP -= (player1[3] / player2[4]) * 60 * multiplier
  else:
    multiplier = Type_multiplier(player1[1],player2[1])

    print('multi : ',multiplier)

    p1_HP -= (player2[3] / player1[4]) * 60 * multiplier
  print(f'{player1[0]} HP: {p1_HP}')
  print(f'{player2[0]} HP: {p2_HP}')
```

```python
#Z_attack
def z_attack(player):
  global p1_HP, p2_HP, z_flag_p1, z_flag_p2
  if player == player1:
    if z_flag_p1 == False:
      attack(player1)
      p2_HP -= 50
      z_flag_p1 = True
    else:
      print("Z Attack Already used, failed to use again")

  else:
    if z_flag_p2 == False:
      attack(player2)
      p1_HP -= 50
      z_flag_p2 = True
    else:
      print("Z Attack Already used, failed to use again")
```
